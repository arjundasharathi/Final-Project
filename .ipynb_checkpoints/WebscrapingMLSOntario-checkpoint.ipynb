```python
House_Listings = []


for x in range(5):

    url = f'https://listing.ca/mls/?..........{x+1}..$'
    response = requests.get(url)
    soup = bs(response.text, "html.parser")
    houses = soup.find_all('div', class_="lc")

    for house in houses:

        try:
            Address = house.find('div', class_="slt_address").text
            Beds = house.find('div', class_="slt_beds").text
            Baths = house.find('div', class_="slt_baths").text
            Price = house.find('div', class_="slt_price").text
            Location = house.find('div', class_="sl_loc").a.text
        

            #print('Latitude = {}, Longitude = {}'.format(location.latitude, location.longitude))


#             print("---------------")
#             print(Address)
#             print(Beds)
#             print(Baths)
#             print(Price)
#             print(Location)
            

            Listings = {}
            
            Listings['Address'] = Address
            Listings['Beds'] = Beds
            Listings['Baths'] = Baths
            Listings['Price'] = Price
            Listings['Location'] = Location
#             Listings['Latitude'] = location.latitude
#             Listings['Longitude'] = location.longitude
            
            House_Listings.append(Listings)
            
            
        except AttributeError as e:
            print("---------------")
            print(e)
    
    
```

```python
DF=pd.DataFrame(House_Listings)
```

```python
DF['Full_Address'] = DF['Address']+', '+DF['Location']+', '+'Canada'
```

```python
DF
```

```python
Houses = pd.DataFrame(House_Listings)
```

```python
lat=[]
long=[]

for x in DF['Full_Address']:

    print (x)

    locator = Nominatim(user_agent='foursquare_agent',timeout=15)
    location = locator.geocode(str(x))
    
    print(location)
    
    if location:
        
        lat.append(location.latitude)
        long.append(location.longitude)
    
    else :
        
        lat.append(None)
        long.append(None)
#         lat = None
#         long = None
    
    
    
    
```

```python
Houses.head()
```

```python
Houses['Latitude'] = lat
Houses['Longitude'] = long
```

```python
Houses=Houses.dropna()
Houses.head()
```

```python
Houses['Price'] = Houses['Price'].str.replace('CAD', '')
Houses['Price'] = Houses['Price'].str.replace('$', '')
Houses['Price'] = Houses['Price'].str.replace(',', '')
Houses['Price'] = Houses['Price'].astype(int)
Houses.head()
```

```python
Houses['Baths'] = Houses['Baths'].str.replace('baths', '')
Houses['Baths'] = Houses['Baths'].astype(int)

```
